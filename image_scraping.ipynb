```python
def extract_info_from_page(page_url, **kwargs):
    info = {}
    result = requests.get(url)
    soup = BeautifulSoup(result.content, 'lxml')
    page_soup =soup.find_all('div', {'class': 'gallery_box_photo'})
    for image_container in page_soup:
        image_container = image_container.a
        image_page_url = get_image_page_url(image_container['href'])
        image_numbers = get_numbers(image_container.img['alt'])
        
        image_page = requests.get(image_page_url)
        image_page_soup = BeautifulSoup(image_page.content, 'lxml')
        image_url = image_page_soup.find('img', {'id': 'mainPicture'})['src']
    
        image_key = re.search(r"(\d+)", image_page_url).group(1)
        info[image_key] = image_numbers

        save_img(image_key, image_url, directory=directory)
    return info
```

```python
import re

def get_image_page_url(a):
    pattern = "'(.+)'"
    return re.search(pattern, a).group(1)

def get_numbers(a):
    pattern = "(\d+)"
    return re.findall(pattern, a)

# new_url = get_image_url(a['href'])
# get_numbers(a.img['alt'])

def save_img(name, image_url, directory):
    with open(directory + name  +".jpg", 'wb') as handle:
            response = requests.get(image_url, stream=True)

            if not response.ok:
                print(response)

            for block in response.iter_content(1024):
                if not block:
                    break
                handle.write(block)
```
